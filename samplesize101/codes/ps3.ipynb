### Q1 : 標準正規分布c.d.fの0.025, 0.05, 0.1, 0.90, 0.95, 0.975地点の値を計算せよ

```python
from scipy.stats import norm
```

```python
### from scratch

### inverse function of normal cdf
def inverse_cdf(p):
    import math
    
    """
    function : 正規分布cdfの逆関数
    input: prob in [0, 1]
    return : 正規分布のvalue
    """
    
    def formula(t):
    #constants
        c0 = 2.515517
        c1 = 0.802853
        c2 = 0.010328
        d0 = 1.432788
        d1 = 0.189269
        d2 = 0.001308  
 
        # Formula
        p = t - ((c2*t + c1)*t + c0) / (((d2*t + d1)*t + d0)*t + 1.0)
        return  p
    
    if (p < 0.5):
        #F^-1(p) = - G^-1(p)
        return -formula( math.sqrt(-2.0*math.log(p)))
 
    else:
        #F^-1(p) = G^-1(1-p)
        return formula( math.sqrt(-2.0*math.log(1-p)))
```

```python
prob_list = [0.025, 0.05, 0.1, 0.90, 0.95, 0.975]
scipy_array = []
scratchnorm_array = []

for p in prob_list:
    scipy_array.append(norm.ppf(p))
    scratchnorm_array.append(inverse_cdf(p))
    
df = pd.DataFrame({'alpha':prob_list , 'scipy':scipy_array,
              'original':scratchnorm_array})
df = df.set_index('alpha')
df['diff'] = abs(df['scipy'] - df['original'])
print(df)
```

```python
## means
mu_0 = 0

## sigma
sigma_1 = sigma_0 =1

## population 1 mean
z = norm.ppf(0.2)
mu_1 = (-2) * z * sigma_1 + mu_0

m = (mu_0 + mu_1)/2
```

### Q4: 検出力曲線を描け

```python
def calculate_beta(n, d, alpha):
    significance_level = round(norm.ppf(alpha/2), 3)
    lower_prob = norm.cdf(significance_level - np.sqrt(n)*d)
    upper_prob = 1 - norm.cdf(-1 * significance_level - np.sqrt(n)*d)
    
    return lower_prob + upper_prob
```

```python
d = np.linspace(-1, 1, 100)
sample_list = [9, 16, 25]

beta_1 = []
beta_2 = []
beta_3 = []

for di in d:
    beta_1.append(calculate_beta(n = sample_list[0], d = di, alpha = 0.05))
    beta_2.append(calculate_beta(n = sample_list[1], d = di, alpha = 0.05))
    beta_3.append(calculate_beta(n = sample_list[2], d = di, alpha = 0.05))
```

```python
plt.figure(figsize = (10,4))
plt.plot(d, beta_1, label = 'sample = {}'.format(sample_list[0]))
plt.plot(d, beta_2, label = 'sample = {}'.format(sample_list[1]))
plt.plot(d, beta_3, label = 'sample = {}'.format(sample_list[2]))
plt.title('power analysis curve')
plt.legend(loc = 4)
;
```

### Q5: z-test関数を定義せよ

### Requirements
- scipy.norm

```python
def _zstat_generic(value1, value2, std_diff, alternative):
    '''generic (normal) z-test to save typing

    can be used as ztest based on summary statistics

    '''
    zstat = (value1 - value2) / std_diff
    if alternative in ['two-sided', '2-sided', '2s']:
        pvalue = (1 - norm.cdf(np.abs(zstat)))*2
    elif alternative in ['larger', 'l']:
        pvalue = 1 - norm.cdf(zstat)
    elif alternative in ['smaller', 's']:
        pvalue = norm.cdf(zstat)
    else:
        raise ValueError('invalid alternative')
    return zstat, pvalue
```

```python
def ztest(x1, x1_var = None, value=0, alternative='two-sided'):
    '''test for mean based on normal distribution, one or two samples

    In the case of two samples, the samples are assumed to be independent.

    Parameters
    ----------
    x1 : array_like, 1-D
        first of the two independent samples
    value : float
        In the one sample case, value is the mean of x1 under the Null
        hypothesis.
    x1_var : standard_deviation
    alternative : string
        The alternative hypothesis, H1, has to be one of the following

           'two-sided': H1: difference in means not equal to value (default)
           'larger' :   H1: difference in means larger than value
           'smaller' :  H1: difference in means smaller than value

    Returns
    -------
    zstat : float
        test statisic
    pvalue : float
        pvalue of the z-test

    '''
    x1 = np.asarray(x1)
    nobs1 = x1.shape[0]
    x1_mean = x1.mean(0)

    if x1_var == None:
        _x1_var = np.sqrt(x1.var())/np.sqrt(nobs1 - 1)
    else:
        _x1_var = x1_var/np.sqrt(nobs1)

    return _zstat_generic(x1_mean, value2 = value, std_diff = _x1_var, alternative = alternative)
```
